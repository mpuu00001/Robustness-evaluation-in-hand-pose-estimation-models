```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
```

# Occlusion Types

```python
types = ['black_circle', 
         'white_circle', 
         'background_coloured_circle',
         'black_circle_5_radius']
```

```python
sample_image_index = 23
sample_image = f"{sample_image_index:08d}.jpg"
folder_path = f'/Users/muxin/PyCharm/organised_FreiHAND/transformed_naked_hands/level_occlusion/'
fig = plt.figure(figsize=(16, 8))
columns, rows = 3, 1

for i in range(1, columns*rows +1):
    image_path = folder_path + f'{str(types[i-1])}/level3/' + sample_image
    image = cv2.imread(image_path)
    fig.add_subplot(rows, columns, i)
    plt.imshow(image)
plt.show()
    
```

```python
folder_path = f'/Users/muxin/PyCharm/organised_FreiHAND/transformed_naked_hands/level_occlusion/{str(types[0])}/'

fig = plt.figure(figsize=(16, 40))
columns, rows = 3, 7
for i in range(1, columns*rows +1):
    image_path = folder_path + f'level{i}/' + sample_image
    image = cv2.imread(image_path)
    fig.add_subplot(rows, columns, i)
    plt.imshow(image)
plt.show()
```

# Read data

```python
folder_path = '/Users/muxin/PyCharm/results/level_occlusion/overall/hand_recognition/'
white_circle = pd.read_csv(f'{folder_path}white_circle.csv')
background_coloured_circle = pd.read_csv(f'{folder_path}background_coloured_circle.csv')
black_circle = pd.read_csv(f'{folder_path}black_circle.csv')
black_circle_5_radius = pd.read_csv(f'{folder_path}black_circle_5_radius.csv')

dfs1 = [white_circle, background_coloured_circle, black_circle, black_circle_5_radius]#, black_rectangle]
lables = ['White circle', 'Background coloured circle', 'Black circle', 'Black circle of 5 radius'] #, 'Black rectangle']
colours = ['green', 'blueviolet', 'blue', 'red']
```

```python
baseline = pd.read_csv('/Users/muxin/PyCharm/results/baseline/overall/hand_recognition/naked_hands.csv')
for i in range(len(dfs1)): 
    dfs1[i] = pd.concat([baseline, dfs1[i]], ignore_index=True)
```

# Hand recognise analysis

```python
fig = plt.subplots(figsize=(12, 6))
plt.xlabel("Occlusion level", fontsize=12, fontweight='bold')
plt.ylabel("Precision", fontsize=12, fontweight='bold')

for i in range(len(dfs1)):
    plt.plot(dfs1[i]['precision'], color=colours[i], label=lables[i])
plt.xticks(range(0, 22))   

x_coordinates = [5, 9, 14, 17]
for x_value in x_coordinates: 
    plt.axvline(x = x_value, color = 'grey', linestyle='--')

plt.legend(loc="upper right", fontsize=12)
# Fixed-size circle occlusion changes the shape of hands sepicallt, 
# resulting the number of recognised hands decreased markablely 
```

```python
fig = plt.subplots(figsize=(12, 6))

for i in range(len(dfs1)):
    plt.plot(dfs1[i]['recall'], color=colours[i], label=lables[i])
plt.xticks(range(0, 22))   

plt.xlabel("Occlusion level", fontsize=12, fontweight='bold')
plt.ylabel("Recall", fontsize=12, fontweight='bold')
plt.xticks(range(0, 22))

x_coordinates = [5, 9, 13, 17]
for x_value in x_coordinates: 
    plt.axvline(x = x_value, color = 'grey', linestyle='--')

plt.legend(loc="upper right", fontsize=12)
```
