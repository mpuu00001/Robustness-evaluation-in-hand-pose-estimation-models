```python
import numpy as np
import argparse
import cv2
def adjust_gamma(image, gamma=1.0):
    # build a lookup table mapping the pixel values [0, 255] to
    # their adjusted gamma values
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")
    # apply gamma correction using the lookup table
    return cv2.LUT(image, table)
```

```python
fig = plt.figure(figsize=(16, 12))
columns, rows = 3, 2
gamma_values = [0.2, 0.5, 1, 2, 8]  #[0.2, 0.5, 1, 2, 5]
titles = ['Strongly underexposed', 'Underexposed', 'Correct exposure', 'Overexposed', 'Strongly overexposed']

for i in range(1, columns*rows +1): 
    if i < 6:
        image_gamma_correct=adjust_gamma(image, gamma_values[i-1])
        fig.add_subplot(rows, columns, i)
        plt.title(titles[i-1])
        plt.imshow(image_gamma_correct)
plt.show()
```
